## 1. Problem Statement

Banks offer 24+ financial products. Mass-marketing all products to all customers is:
- **Expensive**: SMS/email campaigns cost per contact
- **Annoying**: Irrelevant offers reduce customer satisfaction
- **Inefficient**: Typical conversion rate <1% for untargeted campaigns

**Solution**: Predict which specific product(s) each customer is most likely to adopt → target only high-probability customers → higher ROI per campaign.

---

## 2. Data & Approach

### Dataset
- **Source**: Santander Bank (Kaggle competition)
- **Size**: 13.6M rows × 48 columns
- **Period**: 18 months (Jan 2015 – May 2016)
- **Customers**: ~935K unique
- **Products**: 24 financial products (binary ownership flags)

### Target Variable
For each customer-month pair: which products changed from 0 → 1 (newly added).

### Approach
1. **One model per product**: Binary classification for each of the 24 products
2. **Time-based validation**: Train on months < May 2016, validate on May 2016
3. **Model selection**: Best of LightGBM vs XGBoost per product
4. **Explainability**: SHAP values for business stakeholder communication

## 3. Feature Engineering

### Feature Groups (total: ~80+ features)

| Group | Examples | Rationale |
|-------|---------|----------|
| **Demographics** | age, income (log), gender, country | Customer profile baseline |
| **Tenure** | months_since_join, is_new_customer, tenure_bin | Customer lifecycle stage |
| **Product Portfolio** | total_products, product_diversity, payroll_bundle, n_accounts, n_investments | Cross-sell potential |
| **Lag Features** | 24× product_lag1 (did they have it last month?) | Most predictive — recent behavior |
| **Velocity** | products_gained/lost, net_change, momentum | Behavioral trajectory |
| **Temporal** | month_sin/cos, is_q1, is_post_summer | Seasonality effects |
| **Categorical** | segment, channel, province (encoded) | Customer acquisition context |

### Key Insight: Lag Features Dominate
The strongest predictor of "will a customer add product X?" is whether they already have related products. The payroll bundle (Payroll + Pensions + Direct Debit, correlation 0.78-0.95) acts as a gateway — customers who have one typically get the others.

## 5. Business Impact Estimation

### Scenario: Credit Card Cross-sell Campaign

**Without model (mass campaign):**
- Target: All 935K customers
- Conversion rate: ~0.5% (industry average)
- Cost per SMS: €0.03
- Campaign cost: €28,050
- Expected conversions: 4,675
- **Cost per conversion: €6.00**

**With model (top 10% targeting):**
- Target: Top 93.5K customers by model score
- Expected precision in top 10%: ~3-5% (6-10× improvement)
- Campaign cost: €2,805
- Expected conversions: ~3,740 (80% of total in 10% of audience)
- **Cost per conversion: €0.75**

**Impact: ~8× reduction in cost per conversion, 90% reduction in campaign cost**

## 6. Limitations & Next Steps

### Limitations
- No transaction-level data (only monthly snapshots)
- No campaign response history (can't model offer fatigue)
- Some products have very few positive examples (<50)
- Single-country data (Spain) — patterns may not transfer

### Next Steps
1. **Production deployment**: Schedule monthly batch scoring + real-time API
2. **A/B testing**: Validate model-driven campaigns vs. rule-based targeting
3. **Sequence modeling**: Use LSTMs/Transformers for product adoption paths
4. **Reinforcement learning**: Optimize offer timing and channel selection
5. **Transfer to real bank data**: Apply framework to AlternatifBank's CRM data
